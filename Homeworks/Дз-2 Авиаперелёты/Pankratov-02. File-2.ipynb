### Определим Timezome для каждого пропущенного значения

```python
#!pip install timezonefinder
#!pip install geopy
```

```python
from geopy.geocoders import Nominatim
from timezonefinder import TimezoneFinder

def Timezone(lad):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.geocode(lad)
    obj = TimezoneFinder()
    try:
        result = obj.timezone_at(lng=location.longitude, lat=location.latitude)
    except:
        return np.NaN
    return result
```

### P.s. следующая функция работает немного долго! (у меня заняло ~7-8 минут)
#### Идем по всему фрейму и там где Timezone_database - не указана, указываем, используя, написанную выше функцию

```python
for i in airport_row.index:
    if airport_row.Tz_database_time_zone[i] == '\\N':
        airport_row.iloc[i,-3] = Timezone(airport_row.iloc[i,2])
        
airport_row[airport_row.Tz_database_time_zone=='\\N']
```
